**Question no 2:What are the key statistical properties and distributions of the numerical 
columns?**â€¢ Analyze columns such as Age, Fare, and Parch using summary statistics and 
visualizations.
â€¢ Plot histograms, boxplots, and KDE plots to understand distributions and detect 
outliers.

```python
#boxplot
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.xlabel("Fare")
plt.show()
```

```python
#Distplot
sns.histplot(df['Age'],kde=True, bins=20, color='red')
plt.title("Age Distribution with KDE")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Heatmap
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Histogram
df['Age'].plot(kind='hist', bins=20, color='red', alpha=0.7)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

**Question no 4:What is the frequency distribution of categorical columns?** â€¢ Explore Pclass, Sex, Embarked, and Survived using value_counts().
â€¢ Visualize the distributions using countplots and pie charts.

```python
#Pie Chart
survival_counts = df['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=['coral', 'skyblue'])
plt.title("Survival Proportion")
plt.show()
```

```python
#Numerical Data
sns.histplot(df['Fare'], kde=True, color='blue')
plt.title("Distribution of Fare")
plt.show()
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
#Categorical Data
sns.countplot(x='Pclass', data=df, palette='viridis')
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
df['Embarked'].value_counts().plot(kind='bar', color='orange')
plt.title("Embarked Port Count")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()
```

**4. Question: How are variables related to each other?**
Analyze the relationship between:
**â–ª Numerical-Numerical:** Age vs Fare (scatter plot and correlation).
**â–ª Categorical-Categorical:** Pclass vs Survived (grouped bar chart).
**â–ª Numerical-Categorical:** Age distribution across survival status (boxplot).

```python
#Numerical-Numerical
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title("Scatterplot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
#Categorical-Categorical
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
#Numerical-Categorical
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Boxplot of Age by Survival Status")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()
```
